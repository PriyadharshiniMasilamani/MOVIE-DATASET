```python
#plt.figure(figsize=(10, 6))
x,y = 'Year of Release', 'Box Office Success'
df1 = dataset.groupby(x)[y].value_counts(normalize=True)
df1 = df1.mul(100)
df1 = df1.rename('percent').reset_index()

g = sns.catplot(x=x,y='percent',hue=y,kind='bar',data=df1,height=18,aspect=35/15)
g

for p in g.ax.patches:
    txt = str(p.get_height().round(2)) + '%'
    txt_x = p.get_x()
    txt_y = p.get_height()
    g.ax.text(txt_x,txt_y,txt,fontsize=23)
    
g.ax.set_xlabel(x, fontsize=30)
g.ax.set_ylabel('Percentage', fontsize=30)

# Increase font size of tick labels
g.ax.tick_params(axis='x', labelsize=20)
g.ax.tick_params(axis='y', labelsize=20)
```

From this above graph,

i) In the year 2011, the box office success are good/best which is about 24.24% and followed by 2010 which is 20.25% and 2012 which is 19.57% and there is no movies are good in the year 2015.

ii) In 2012,most of the movies are above average in the box office success which is about 36.96% and followed by 2014 which is 26.73% and 2016 which is 23.42% and so on.

ii) In 2016,most of the movies are average in the box offfice success which is about 48.65%  and followed by 2010 which is 44.3% and 2021 which is 42.42% and so on.

iii) In 2015,most of the movies are below average in the box office success which is about 57.14% and followed by 2020 which is about 51.28% and 2017 which is 46.1% and so on.

* In 2017,the number of movies releases are more.
